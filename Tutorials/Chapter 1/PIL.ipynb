```python
from PIL import Image
```

```python
import wget
```

```python
width = 500
height = 500
```

```python
path = "./images/"
url = "https://picsum.photos/{}/{}".format(width, height)
```

```python
im_path = wget.download(url, out=path)
```

```python
pil_im = Image.open(im_path)
```

```python
pil_im
```

```python
box = (100, 100, 400, 400)
region = pil_im.crop(box)
```

```python
region
```

```python
region = region.transpose(Image.ROTATE_180)
pil_im.paste(region, box)
```

```python
pil_im
```

```python
out = pil_im.resize((128, 128))
```

```python
out
```

```python
out = pil_im.rotate(45)
```

```python
out
```

```python
pil_im_grayscale = pil_im.convert('L')
```
