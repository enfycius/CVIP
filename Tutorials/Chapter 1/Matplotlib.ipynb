```python
import os
```

```python
def get_imlist(path):
    return [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
path = "./images/"
```

```python
im_path = get_imlist(path)[0]
```

```python
from PIL import Image
from pylab import *
```

```python
im = array(Image.open(im_path))
```

```python
imshow(im)
```

```python
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]
```

```python
imshow(im)
plot(x, y, 'r*')
```

```python
imshow(im)
plot(x[:2], y[:2])
```

```python
imshow(im)

plot(x, y, 'r*')
plot(x[:2], y[:2])
```
