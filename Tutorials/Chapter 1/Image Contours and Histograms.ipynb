```python
from PIL import Image
from pylab import *
```

```python
import os
```

```python
def get_imlist(path):
    return [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
path = "./images/"
```

```python
im_path = get_imlist(path)[0]
```

```python
im = array(Image.open(im_path).convert('L'))
```

```python
contour(im, origin='image')
axis('equal')
axis('off')
```

```python
hist(im.flatten(), 128)
```
